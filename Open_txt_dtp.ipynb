```python
def save_csv_world(labels_path, output_dir):
    damage_list = os.listdir(labels_path)
    damage_data = pd.DataFrame({'File_name':[], 'Frame_number':[], 'File_time(sec)':[], 'Damage_class':[], 'X_center':[], 'Y_center':[], 'Width':[], 'Height':[], 'Confidence':[], 'Glonass':[], 'Speed':[], 'GR_time':[]})
    s_index=0
    for fil in damage_list:
        #frame_number = int(fil.split('_')[1])
        file_name = fil
        print(fil, s_index)
        #sec = frame_number//30+1
        with open (labels_path+fil, 'r') as nf:
            l = nf.readlines()
            if l:
                for i in l:
                    damage_data.loc[s_index, 'File_name'] = fil
                    #damage_data.loc[s_index, 'Frame_number'] =  frame_number
                    #damage_data.loc[s_index, 'File_time(sec)'] =  sec
                    damage_data.loc[s_index, 'Damage_class'] =  str(i.strip().split()[0])
                    damage_data.loc[s_index, 'X_center'] =  float(i.strip().split()[1])
                    damage_data.loc[s_index, 'Y_center'] =  float(i.strip().split()[2])
                    damage_data.loc[s_index, 'Width'] =  float(i.strip().split()[3])
                    damage_data.loc[s_index, 'Height'] =  float(i.strip().split()[4])
                    s_index +=1
    #damage_data.sort_values('Frame_number', inplace=True)
    damage_data.to_csv(output_dir+'Barriers_add.csv', encoding="utf-8-sig", index=False)
    return damage_data
```
